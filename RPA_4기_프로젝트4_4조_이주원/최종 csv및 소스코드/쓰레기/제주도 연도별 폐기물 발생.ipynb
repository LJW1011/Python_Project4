```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

plt.rcParams['font.family']='Malgun Gothic'
plt.rcParams['axes.unicode_minus']=False
```

```python
df = pd.read_csv('수집된 데이터/쓰레기 관련/제주도 연도별 폐기물 발생현황.csv',encoding='cp949')
```

```python
df["소각쓰레기량(%)"].unique()
```

```python
df["합계"].unique()
```

```python
df["재활용쓰레기량(%)"] = df.iloc[:,8].str.replace("%","").astype("float64")
df["매립쓰레기량(%)"] = df.iloc[:,6].str.replace("%","").astype("float64")
df["소각쓰레기량(%)"] = df.iloc[:,4].str.replace("%","").astype("float64")
df["합계"] = df.iloc[:,2].str.replace(",","").astype("float64")
```

```python
plt.rcParams['figure.figsize']=(12,5)
fig, axes = plt.subplots(1,3, facecolor='snow')
sns.barplot(df, x="연도별", y = "소각쓰레기량(톤/일)",hue = "제주도시별", ax = axes[1]).set(title = '연도별 제주도 소각쓰레기 배출량')
axes[1].grid(axis='y', ls=':',color ='k', alpha = 0.3)
sns.barplot(df, x="연도별", y = "매립쓰레기량(톤/일)",hue = "제주도시별", ax = axes[2]).set(title = '연도별 제주도 매립쓰레기 배출량')
axes[2].grid(axis='y', ls=':',color ='k', alpha = 0.3)
sns.barplot(df, x="연도별", y = "재활용쓰레기량(톤/일)",hue = "제주도시별", ax = axes[0]).set(title = '연도별 제주도 재활용쓰레기 배출량')
axes[0].grid(axis='y', ls=':',color ='k', alpha = 0.3)
fig.suptitle("제주도 종합쓰레기 배출량",size =20,color = "darkblue",fontweight ="bold")
fig.tight_layout()
plt.show()
```

```python
plt.rcParams['figure.figsize']=(12,5)
fig, axes = plt.subplots(1,3, facecolor='snow')
sns.barplot(df, x="연도별", y = "소각쓰레기량(%)",hue = "제주도시별", ax = axes[1]).set(title = '연도별 제주도 소각쓰레기 배출비율')
axes[1].grid(axis='y', ls=':',color ='k', alpha = 0.3)
sns.barplot(df, x="연도별", y = "매립쓰레기량(%)",hue = "제주도시별", ax = axes[2]).set(title = '연도별 제주도 매립쓰레기 배출비율')
axes[2].grid(axis='y', ls=':',color ='k', alpha = 0.3)
sns.barplot(df, x="연도별", y = "재활용쓰레기량(%)",hue = "제주도시별", ax = axes[0]).set(title = '연도별 제주도 재활용쓰레기 배출비율')
axes[0].grid(axis='y', ls=':',color ='k', alpha = 0.3)
axes[0].set_yticks(range(0,101,10))
fig.suptitle("제주도 종합쓰레기 배출비율",size =20,color = "darkblue",fontweight ="bold")
fig.tight_layout()
plt.show()
```
