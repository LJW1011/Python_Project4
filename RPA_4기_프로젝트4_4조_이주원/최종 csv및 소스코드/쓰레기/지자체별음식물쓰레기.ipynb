```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

plt.rcParams['font.family']='Malgun Gothic'
plt.rcParams['axes.unicode_minus']=False
```

```python
df = pd.read_csv('수집된 데이터/쓰레기 관련/지자체별음식물쓰레기배출내역.csv')
```

```python
df.shape
```

```python
df.isnull().sum()
```

```python
df = df.drop(columns=['apt_cd','apt_nm','locgov_relm_cd'])
```

```python
df.info()
```

```python
df.columns=['배출년도','배출월','배출일','배출요일','시도명','시군구','배출량','배출량률','배출횟수','배출횟수율']
```

```python
df.head()
```

```python
df['배출량률'] = round(df['배출량률'],1)
df['배출횟수율'] = round(df['배출횟수율'],1)
```

```python
df_2018 = df[df['배출년도']==2018].sort_values('배출량', ascending=False)
df_2018
```

```python
df_2019 = df[df['배출년도']==2019].sort_values('배출량', ascending=False)
df_2019
```

```python
df_2020 = df[df['배출년도']==2020].sort_values('배출량', ascending=False)
df_2020
```

```python
df_sido_2018 = df_2018.groupby("시도명")[["배출량","배출횟수"]].mean().sort_values("배출량",ascending= False)
df_sido_2018
```

```python
df_sido_2019 = df_2019.groupby("시도명")[["배출량","배출횟수"]].mean().sort_values("배출량",ascending= False)
df_sido_2019
```

```python
df_sido_2020 = df_2020.groupby("시도명")[["배출량","배출횟수"]].mean().sort_values("배출량",ascending= False)
df_sido_2020
```

```python
df_18 = df_2018.pivot_table(index = "시도명", values= ["배출량","배출횟수"] ,aggfunc="mean")
df_18 = df_18.sort_values("배출량",ascending =False)
df_18_no = df_18.drop("제주특별자치도")
df_19 = df_2019.pivot_table(index = "시도명", values= ["배출량","배출횟수"] ,aggfunc="mean")
df_19 = df_19.sort_values("배출량",ascending =False)
df_19_no = df_19.drop("제주특별자치도")
df_20 = df_2020.pivot_table(index = "시도명", values= ["배출량","배출횟수"] ,aggfunc="mean")
df_20 = df_20.sort_values("배출량",ascending =False)
df_20_no = df_20.drop("제주특별자치도")
```

```python
df_18_no.reset_index(inplace = True)
df_18.reset_index(inplace = True)
df_19_no.reset_index(inplace = True)
df_19.reset_index(inplace = True)
df_20_no.reset_index(inplace = True)
df_20.reset_index(inplace = True)
```

```python
cc2018={"시도명":["내륙"],"배출량": [sum(df_18_no["배출량"])/len(df_18_no["배출량"])], "배출횟수": [sum(df_18_no["배출횟수"])/len(df_18_no["배출횟수"])]}
cc18 = pd.DataFrame(cc2018)
cc2019={"시도명":["내륙"],"배출량": [sum(df_19_no["배출량"])/len(df_19_no["배출량"])], "배출횟수": [sum(df_19_no["배출횟수"])/len(df_19_no["배출횟수"])]}
cc19 = pd.DataFrame(cc2019)
cc2020={"시도명":["내륙"],"배출량": [sum(df_20_no["배출량"])/len(df_20_no["배출량"])], "배출횟수": [sum(df_20_no["배출횟수"])/len(df_20_no["배출횟수"])]}
cc20 = pd.DataFrame(cc2020)
```

```python
df_18 = pd.concat([df_18,cc18],ignore_index= True)
df_19 = pd.concat([df_19,cc19],ignore_index= True)
df_20 = pd.concat([df_20,cc20],ignore_index= True)
```

```python
df_jj_18 = df_18.loc[[1,16]]
df_jj_18.index = df_jj_18["시도명"]
df_jj_18.drop(columns="시도명",inplace=True)
df_jj_19 = df_19.loc[[1,16]]
df_jj_19.index = df_jj_19["시도명"]
df_jj_19.drop(columns="시도명",inplace=True)
df_jj_20 = df_20.loc[[1,16]]
df_jj_20.index = df_jj_20["시도명"]
df_jj_20.drop(columns="시도명",inplace=True)
```

```python
df_jj_20
```

```python
df_jj_18.reset_index(inplace = True)
df_jj_18 = df_jj_18.rename(index = {0:"2018년", 1:"2018년"})
df_jj_19.reset_index(inplace = True)
df_jj_19 = df_jj_19.rename(index = {0:"2019년", 1:"2019년"})
df_jj_20.reset_index(inplace = True)
df_jj_20 = df_jj_20.rename(index = {0:"2020년", 1:"2020년"})
```

```python
df_jj_20
```

```python
df_jj = pd.concat([df_jj_18,df_jj_19,df_jj_20],ignore_index= True)
df_jj = df_jj.rename(index = {0:"2018년", 1:"2018년",2:"2019년",3:"2019년",4:"2020년",5:"2020년"})
df_jj
```

```python
df_jeju = df[df['시도명']=="제주특별자치도"]
df_jeju
```

```python
df_jejusi = df_jeju.groupby(["배출년도","시군구"], as_index = False).agg(평균배출량 = ('배출량', 'mean'),평균배출횟수 = ('배출횟수', 'mean'))
```

```python
fig, axes = plt.subplots(1,2, figsize=(10,5), facecolor='snow')
sns.barplot(data = df_jejusi, x = "배출년도", y = '평균배출량', hue = "시군구", ax = axes[0], palette ={'제주시':'gray', '서귀포시':'darkorange'}
           ).set(title = '연도별 음식물쓰레기 배출량')
axes[0].grid(axis='y', ls=':',color ='k', alpha = 0.3)
axes[0].legend(loc=(0.4, 0.87))
sns.barplot(data = df_jejusi, x = "배출년도", y = '평균배출횟수', hue = "시군구", ax = axes[1],palette ={'제주시':'gray', '서귀포시':'darkorange'}
           ).set(title = '연도별 음식물쓰레기 배출횟수')
axes[1].grid(axis='y', ls='--', color='r',alpha = 0.3)
axes[1].legend(loc=(0.05, 0.87))
fig.suptitle("제주도 음식물쓰레기 배출",size =20,color = "darkred",fontweight ="bold")
fig.tight_layout()
plt.show()
```
